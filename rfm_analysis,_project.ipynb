```python
rfm = pd.concat([recency_df, freq_df, monetary_df],  axis=1) #rfm dataframe
```

```python
rfm.head()
```

```python
rfm["RecencyScore"] = pd.qcut(rfm["Recency"], 5, labels = [5, 4 , 3, 2, 1]) 
#sorted and calculated 5 quantile values for Recency  and labeled (descending, this is important!)
```

```python
rfm.head()
```

```python
#rfm["FrequencyScore"] = pd.qcut(rfm['Frequency'], 5, labels = [1, 2, 3, 4, 5]) 
#rank to take first value of a quantile in order to handle error
```

```python
rfm["FrequencyScore"]= pd.qcut(rfm["Frequency"].rank(method="first"),5, labels=[1,2,3,4,5])
#sorted and calculated 5 quantile values for Frequency  and labeled (ascending, this is important!)
```

```python
rfm.head()
```

```python
rfm["MonetaryScore"] = pd.qcut(rfm['Monetary'], 5, labels = [1, 2, 3, 4, 5])
#sorted and calculated 5 quantile values for Monetary  and labeled (ascending, this is important!)
```

```python
rfm.head()
```

```python
(rfm['RecencyScore'].astype(str) + 
 rfm['FrequencyScore'].astype(str) + 
 rfm['MonetaryScore'].astype(str)).head()
```

```python
# rfm skorları kategorik değere dönüştürülüp df'e eklendi
rfm["RFM_SCORE"] = (rfm['RecencyScore'].astype(str) + 
                    rfm['FrequencyScore'].astype(str) + 
                    rfm['MonetaryScore'].astype(str)) # RFM score calculation
```

```python
# RFM Map via Regular Expressions
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
#we'll use only Recency and Frequency scores to calculate RFM Scores
```

```python
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)# replate Segment values with Regex map
```

# Comments

I chose a segment that brings a lot of money and therefore should not be lost (Can't loose), one is very loyal to us (Loyal Customers), and one is unstable, so we need to be careful. (Need attention)

### Can't loose
- We have 64 customers in this group.
- Despite the small number of customers, it seems to be very important when considering the income from each person. (The median is 2236 pounds, and this median value is the second highest figure when considering the average income earned by people in the segment.)
- Considering the number of customers, the earned income is considered to be truly memorable. Even when comparing with the Champions segment, about 4% of revenues from Champions are generated by 10% of people.
- Considering the median recency values ​​(134) together with the frequency values, the monetary values ​​seem to be extremely high, i.e. the number of people is quite low (64), but the median income per person is extremely high (2236), I conclude from this ; If we increase the number of customers in this segment, its contribution to our revenues will be high. For this, we should remind ourselves frequently, we should organize promotions according to the profiles of this segment and make the friends of this segment as our customers.
- We should try to include customers with the same profile in this segment by making very detailed profile analysis of the customers in this segment.

### Loyal Customers
- We have 820 customers in the segment.
- Considering that one of our 5 customers is in the obu segment and approximately 25% of the revenues (£ 2347570.48) belongs to these customers, the importance of the segment will be clearly seen.
- The mean and median values ​​of the recency values ​​are very good compared to the other segments. (median 28)
- Frequency values ​​are about 6, a good result compared to other segments
- When the monetary values ​​are examined, the mean is 2862, the median is 1737. There is a positive twist here in the monetary distribution.
- Here I have a general comment about loyal customers. A general rule of thumb is often mentioned in Marketing and Business matters, which is: retaining existing, acquired customers, is worth more than acquiring new customers. In practice, it is always important, although not as obvious as in this example. Given that 25% of the revenue in this example is derived from these customers, they must be supported by special promotions, offers, discounts, recommendations and gifts to keep them retained.

### Need Attention
- There are 184 customers in the segment.
- The revenue generated is £ 64586.21 and 2% of the total revenue, so this segment does not think it is necessary to overthrow their customers.
- The recency median is 50, which is an important reason for the very low RFM score.
- Although the revenue earned from this segment is quite low, the number of 184 customers is not a small value and interesting sms can be sent to increase the recency scores. In order to enable them to visit more and buy products from our site, the products they bought before and the products that were sold together and the products that the customer preferred together can be recommended.

In fact, an association rules analysis should be done for customers in all segments in general, and special association rules should be examined by extracting special association rules for the profiles in the segments, and I think that this process should be automated to provide sustainable benefits.
